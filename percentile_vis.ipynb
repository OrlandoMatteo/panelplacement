```python
import numpy as np
import pandas as pd
```

```python
output_traces=pd.read_csv("traces2/area_suit.csv")
```

```python
#output_traces.head(15)
```

```python
percentile=[]
for index,row in output_traces.iloc[:,13:].iterrows():
    percentile.append(np.percentile(np.asarray(row.fillna(0)),75))
```

```python
percentile
```

```python
pd.DataFrame({'x':output_traces.x,'y':output_traces.y,'percentile':percentile}).to_csv('percentile_area.csv')
```
